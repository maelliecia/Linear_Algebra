### Problem 1

```python
# PROBLEM 1
import numpy as np
u = np.array([1,-2,4])
v = np.array([3,5,1])
w = np.array([2,1,-3])

print("PROBLEM 1\n")
print("u = ", u)
print("v = ", v)
print("w = ", w)

# 1.a
print("\na. 3u - 2v = ", 3*u-2*v)

# 1.b
print("b. 5u + 3v - 4w = ", 5*u+3*v-4*w)

#1.c
print("\nc. u.v,v.w,u.w = ", u.dot(v),',',v.dot(w),',',u.dot(w))

#1.d
print("d. ||u||,||v||,||w|| = ", np.sqrt(u.dot(u)),',', np.sqrt(v.dot(v)),',',np.sqrt(w.dot(w)))

#1.e
print("\ne. cos 0:",u.dot(v)/(np.linalg.norm(u)*np.linalg.norm(v)))

#1.f
print("f. cos a:",v.dot(w)/(np.linalg.norm(v)*np.linalg.norm(w)))

#1.g
print("g. cos B:",u.dot(w)/(np.linalg.norm(u)*np.linalg.norm(w)))

#1.h
print("\nh. d(u,v) = ", np.linalg.norm(u-v))

#1.i
print("i. d(w,v) = ", np.linalg.norm(w-v))

#1.j
print("j. d(u,w) = ", np.linalg.norm(u-w))

#1.k
print("\nk. proj(u,v) = ",u.dot(v)*v/(np.linalg.norm(v)**2) )
print("k. proj(u,v) = ",[-9/35, -15/35, -3/35] )

#1.l
print("\nl. proj(w,v) = ",w.dot(v)*v/(np.linalg.norm(v)**2) )
print("l. proj(w,v) = ",[24/35, 40/35, 8/35])

#1.m
print("\nm. proj(u,w) = ",u.dot(w)*w/(np.linalg.norm(w)**2) )
print("m. proj(u,w) = ",[-24/14, -12/14, 36/14] )
```
